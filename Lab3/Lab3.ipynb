```python
import torch
import torchvision.transforms as transforms
import torchvision.transforms.functional as F
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = Image.open("images\dry-maple-leaves-grayscale.webp").convert("RGB")

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Task 1.1:
scaled_img = F.resize(img, size=[int(img.height*1.5), int(img.width*1.5)])

plt.imshow(scaled_img)
plt.title("Scaled Image")
plt.axis("off")
plt.show()
```

```python
# Task 1.2:
rotated_img = F.rotate(img, angle=120)

plt.imshow(rotated_img)
plt.title("Rotated 120 Degrees")
plt.axis("off")
plt.show()
```

```python
# Task 1.3:
sheared_img = F.affine(
    img,
    angle=0,
    translate=(0, 0),
    scale=1,
    shear=30   # shear angle
)

plt.imshow(sheared_img)
plt.title("Sheared Image")
plt.axis("off")
plt.show()
```

```python
to_tensor = transforms.ToTensor()
img_tensor = to_tensor(img)
```

```python
# Task 2.1:
negative = 1 - img_tensor

plt.imshow(negative.permute(1,2,0))
plt.title("Negative Image")
plt.axis("off")
plt.show()
```

```python
# Task 2.2:
c = 1 / torch.log(torch.tensor(2.0))
log_img = c * torch.log(1 + img_tensor)

plt.imshow(log_img.permute(1,2,0))
plt.title("Log Transformation")
plt.axis("off")
plt.show()
```

```python
# Task 2.3:
gamma = 1.5
power_img = img_tensor ** gamma

plt.imshow(power_img.permute(1,2,0))
plt.title("Power Law (Gamma=1.5)")
plt.axis("off")
plt.show()
```
