```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
image = data.chelsea().astype(np.float32) / 255.0

sigma = 2
amount = 1.5

blurred = cv2.GaussianBlur(image, (0, 0), sigmaX=sigma, sigmaY=sigma)
sharpened = np.clip(image + amount * (image - blurred), 0, 1)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")
ax[1].imshow(blurred)
ax[1].set_title("Blurred Image")
ax[1].axis("off")
ax[2].imshow(sharpened)
ax[2].set_title("Sharpened (Unsharp Masking)")
ax[2].axis("off")
plt.show()
```

```python
# Task 1
image = data.camera()

kernel = np.ones((7, 7), np.float32) / 49
smoothed = cv2.filter2D(image, -1, kernel)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(smoothed, cmap='gray')
plt.title('7x7 Box Filter')
plt.axis('off')
plt.show()
```

```python
# Task 2
image = data.camera()

blur5 = cv2.GaussianBlur(image, (5, 5), 0)
blur21 = cv2.GaussianBlur(image, (21, 21), 0)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(blur5, cmap='gray')
plt.title('5x5 Gaussian Filter')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(blur21, cmap='gray')
plt.title('21x21 Gaussian Filter')
plt.axis('off')
plt.show()
```

```python
# Task 3
image = data.camera()
image_float = image.astype(np.float32) / 255.0

laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)
sharpened = np.clip(image_float - laplacian, 0, 1)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image_float, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian Filter')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(sharpened, cmap='gray')
plt.title('Sharpened Image')
plt.axis('off')
plt.show()
```
